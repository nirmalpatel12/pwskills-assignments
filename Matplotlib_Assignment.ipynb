```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
#plot a scatter plot.
```

```python
'''Displays individual data points on a two-dimensional graph to identify patterns or relationships.'''
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x,y,c='b')
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title('scatter plot for data')
plt.show()
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
```

```python
'''The subplot() function in Matplotlib is used to create subplots within a single figure. It allows you to arrange multiple plots in a grid, specifying the number of rows and columns.'''
```

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])  
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(4,4,1)
plt.plot(x1,y1,c='pink')
plt.title('line 1')
plt.xlabel('x1')
plt.ylabel('y1')


plt.subplot(4,4,2)
plt.plot(x1,y2,c='y')
plt.title('line 2')
plt.xlabel('x2')
plt.ylabel('y2')

plt.subplot(4,4,3)
plt.plot(x1,y3,c='r')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('line 3')

plt.subplot(4,4,4)
plt.plot(x1,y4,c='b')
plt.title('line 4')
plt.xlabel('x2')
plt.ylabel('y2')


plt.tight_layout()
plt.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```

```python
'''A bar plot is a graphical representation of categorical data with rectangular bars. Each bar's length or height corresponds to the frequency or value of the category it represents. Bar plots are commonly used to compare the values of different categories or to show the distribution of a single categorical variable.'''
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,6))
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('profit analysis')
plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```

```python
'''A box plot (or box-and-whisker plot) is a statistical visualization that provides a summary of the distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers. Box plots are particularly useful for comparing the distribution of different groups of data and identifying the presence of outliers.'''
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.title('for box1')
plt.show()
```

```python
plt.boxplot(box2)
plt.title('for box2')
plt.show()
```
