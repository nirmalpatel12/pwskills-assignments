```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
#"timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
```

```python
data=sns.load_dataset('fmri')
```

```python
data
```

```python
sns.lineplot(x=data['timepoint'],y=data['signal'])
```

```python
sns.lineplot(x=data['timepoint'],y=data['signal'],hue=data['region'])
```

```python
sns.lineplot(x=data['timepoint'],y=data['signal'],hue=data['event'])
```

```python
sns.lineplot(x=data['timepoint'],y=data['signal'],hue=data['region'],style=data['event'])
```

```python
sns.lineplot(x=data['timepoint'],y=data['signal'],hue=data['subject'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with hue="species"
sns.pairplot(iris, hue="species")

# Show the plot
plt.show()
```
